```python
# 1. 판다스
import pandas as pd
```

```python
# 1) 시리즈
#시리즈 클래스는 1차원 배열의 값(values)에 각 값에 대응되는 인덱스를 부여할 수 있는 구조를 가지고 있습니다. 
sr=pd.Series([17000, 18000, 1000, 5000],
             index=["피자", "치킨", "콜라", "맥주"])
print("시리즈 출력 : ")
print('-'*15)
print(sr)
```

```python
#값(values)와 인덱스(index)를 출력합니다. 
print('시리즈의 값 :{} '.format(sr.values))
print('시리지의 인덱스 :{} '.format(sr.index))
```

```python
#2. 넘파이
# 수치데이터를 다루는 파이썬 패키지입니다. Numpy의 핵심이라고 불리는 다차원 행렬 자료구조인 ndarray를 통해 벡터 및 행렬을 사용하는 선형대수 계산에서 주로 사용됩니다. 
#Numpy는 편의성 뿐만 아니라, 속도면에서도 순수 파이썬에 비해 압도적으로 빠르다는 장점이 있습니다. 
import numpy as np
```

```python
#1) np.array()
vec=np.array([1,2,3,4,5])
print(vec)
```

```python
#array()안에는 하나의 리스트만 들어가므로 리스트의 리스트를 넣어야한다.
#2차원 배열
mat=np.array([[10,20,30],[60,70,80]])
print(mat)
```

```python
print('vec의 타입 : ',type(vec))
print('mat의 타입 : ', type(mat))
```

```python
print('vec의 축의 개수 : ',vec.ndim) #축의 개수 출력
print('vec의 크기(shape) : ', vec.shape) #크기 출력
```

```python
print('mat의 축의 개수 : ',mat.ndim) #축의 개수 출력
print('mat의 크기(shape) : ', mat.shape) #크기 출력
```

```python
#2) ndarray의 초기화
zero_mat=np.zeros((2,3))
print(zero_mat)
```

```python
#모든 값이 1인 2x3 배열 생성
one_mat=np.ones((2,3))
print(one_mat)
```

```python
#모든 값이 특정 상수인 배열 생성, 이 경우에는 7
same_value_mat=np.full((2,2),7)
print(same_value_mat)
```

```python
#대각선 값이 1이고 나머지 값인 0인 2차원 배열을 생성 (항등행렬)
eye_mat=np.eye(3)
print(eye_mat)
```

```python
#임의의 값으로 채워진 배열 생성
random_mat=np.random.random((2,2))
print(random_mat)
```

```python
#np_arange(n)은 0부터 n-1까지의 값을 가지는 배열을 생성합니다.
range_vec=np.arange(10)
print(range_vec)
```

```python
#np.arange(i,j,k)는 i부터 j-1까지 k씩 증가하는 배열을 생성합니다.
#1부터 9까지 +2씩 적용되는 범위
n=2
range_n_step_vec=np.arange(1,10,n)
print(range_n_step_vec)
```

```python
#np.reshape()는 내부 데이터는 변경하지 않으면서 배열의 구조를 바꿉니다. 0부터 29까지의 숫자를 생성하는 arange(30)을 수행한후, 원소의 개수가
#30개 이므로 5행 6열의 행렬로 변경해본다.
reshape_mat=np.array(np.arange(30)).reshape((5,6))
print(reshape_mat)
```

```python
# 5) Numpy 슬라이싱
#nparray를 통해 만든 다차원 배열은 파이썬의 자료구조인 리스트처럼 슬라이싱(slicing) 기능을 지원합니다. 
#슬라이싱 기능을 사용하여 특정 행이나 열들의 원소들을 접근할 수 있습니다. 
mat=np.array([[1,2,3],[4,5,6]])
print(mat) 
```

```python
slicing_mat=mat[:,1]
print(slicing_mat)
```

```python
# 6) Numpy 정수 인덱싱(integer indexing)
#슬라이싱을 사용하면 배열로부터 부분 배열을 추출할 수 있지만, 연속적이지 않은 원소를 배열을 만들 경우에는 슬라이싱으로는 만들수 없다.
# ex. 2행2열의 원소와 5행5열의 원소를 뽑아서 하나의 배열로 만들고자 하는 경우, 이 경우에는 인덱싱을 사용하여 배열을 구성할 수 있다. 
mat=np.array([[1,2],[4,5],[7,8]])
print(mat)
```

```python
print(mat[1,0])
```

```python
indexing_mat=mat[[2,1],[0,1]]
print(indexing_mat)
```

```python
#numpy 연산
#numpy를 사용하면 배열간 연산을 쉽게 수행할 수 있다. 덧셈, 뺄셈, 곱셈, 나눗셈을 위해서는 연산자(*,-,/,+)를 사용할 수 있으며 또는 
#np.add(),np.subtract(), np.multiply(), np.divide()를 사용할 수도 있다. 

x=np.array([1,2,3])
y=np.array([4,5,6])
```

```python
#result=np.add(x,y)와 동일
result=x+y
print(result)
```

```python
# result=np.subtract(x,y)
result=x-y
print(result)
```

```python
#result=np.multiply(result,x)
result=result*x
print(result)
```

```python
#result=np.divide(result,x)와 동일
result=result/x
print(result)
```

```python
#위에서 *을 통해 수행한 것은 요소별 곱이다. Numpy에서 벡터와 행렬곱을 위해서는 dot()을 사용해야 한다.
mat1=np.array([[1,2],[3,4]])
mat2=np.array([[5,6],[7,8]])
mat3=np.dot(mat1,mat2)
print(mat3)
```

```python
#3. matplotlib
#데이터를 차트나 슬롯으로 시각화하는 패키지이다. 데이터 분석에서 Matlplotlib은 데이터 분석 이전에 데이터 이해를 위한 시각화나, 데이터 분석 후에 결과를 시각화
#하기 위해서 사용된다. 아나콘다 설치할 경우 포함되어있음 

import matplotlib.pyplot as plt
```

```python
# 1) 라인 플롯 그리기
#plot은 라인 플롯을 그리느 기능을 수행한다. plot()에 x축과 y축의 값을 기재하고 그림ㅇ르 표시하는 show()를 통해 시각화해보자. 
#title('제목')을 통해 제목을 지정할 수 있다. 주피터 노트북에서는 show를 사용하지 않더라도 자동으로 그래프가 렌더링 되므로 그래프가 시각화되지만 다른 곳에서는 show() 필요

plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.show()
```

```python
#2) 축 레이블 삽입하기 
#x축과 y축 각각에 축이름을 삽입하고 싶다면 xlabel('넣고싶은 축이름')과 ylabel('넣고싶은 축이름')을 사용합니다.
#위의 그래프에서 hours와 score라는 이름을 추가해보자

plt.title('test')
plt.plot([1,2,3,4],[2,4,8,6])
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
#3) 라인추가와 범례 삽입하기
plt.title('students')
plt.plot([1,2,3,4],[3,4,8,6])
plt.plot([1.5,2.5,3.5,4.5],[3,5,8,10])
plt.xlabel('hours')
plt.ylabel('score')
plt.legend(['A Student','B Student'])
plt.show()
```
